```python
# Chatbot with Personality + Polish: contain error handling(try-except), clean code structure , personality , rules ,
# functions-show history, clear history,run chatbot , commands- help,quit,clear,show  , counting messages

!pip install groq

from groq import Groq
# configure api k ey
client = Groq(api_key="Your key here")

count_mssg=0
chat_history=[]
Max_len=10

system_prompt = """You are KrishnaBot — a smart, friendly AI assistant
created by Krishna, a final year CS student and AI developer.

Your personality:
- Friendly and encouraging
- Clear and concise answers
- Always give examples when explaining
- Remember everything user tells you
- If user seems confused — simplify your answer

Your rules:
- Keep answers under 10 lines
- Never say "I am an AI language model"
- Always address user by name if you know it
- End complex answers with "Hope that helps!"

Your expertise:
- Python and Java programming
- Machine Learning and AI
- Career advice for CS students
- General knowledge"""



def show_help():
  print("""
      _____________________________________________________
      These are the commands you can use
                1. help: for showing commands
                2. show: show the full conversation history
                3. clear: clear full history
                4. quit: exit from chat bot
      _____________________________________________________
                """)


# for showing history
def show_history():
    for mssg in chat_history:
      print(f"{mssg["role"]}:{mssg["content"]}")

# for clear the history
def clear_history():
  chat_history.clear()

# for getting ai response keeping chat history
def get_ai_response(prompt):

  chat_history.append({"role":"user","content":prompt})
  if(len(chat_history)>Max_len):
    chat_history.pop(0)

    # try-except for error handling
  try:
    res=client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      messages=[{"role":"system","content":system_prompt}]+chat_history,
      temperature=0.7
       )
    ai_res=res.choices[0].message.content
    chat_history.append({"role":"assistant","content":ai_res})
    return ai_res


  except Exception as e:
    print(f"Error message : {str(e)}")


# the main function which call every upper functions
def run_chatBot():

  global count_mssg

  # Welcome message
  print("""
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
   KrishnaBot — AI Assistant 🤖
   Created by Krishna | Powered by Llama AI
   Type 'help' for commands
   Ask any ques! Type 'Quit' to close
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━""")
  while(True):
    inp=input("you:")
    if(inp.lower()=="quit"):
      print(f"\nKrishnaBot: Goodbye! We had {count_mssg} exchanges today. Keep learning! 💪")
      break
    elif(inp.lower()=="help"):
      show_help()
      continue
    elif(inp.lower()=="clear"):
      clear_history()
      print("HISTORY IS CLEARED")
      continue

    elif(inp.lower()=="show"):
      show_history()
      continue

    print(f"AI:{get_ai_response(inp)}")
    count_mssg+=1

run_chatBot()
# print(count_mssg)




















```
